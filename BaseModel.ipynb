```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
```

```python
import torch
```

```python
# Assuming these are already initialized globally
gpt2_tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
gpt2_model = GPT2LMHeadModel.from_pretrained("gpt2")

```

```python
def generate_gpt2_words_cpu(current_lyrics, end_token="."):
    """Generate k words with GPT-2, stopping early if end_token is found."""

    inputs = gpt2_tokenizer(current_lyrics, return_tensors='pt', truncation=True, max_length=128)
    device = torch.device("cpu")
    gpt2_model.to(device)
    inputs = {key: value.to(device) for key, value in inputs.items()}

    outputs = gpt2_model.generate(
        **inputs,
        max_new_tokens=128,
        do_sample=True,
        top_p=0.95,
        temperature=0.7,
        pad_token_id=gpt2_tokenizer.eos_token_id
    )

    new_words = gpt2_tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

    return new_words

```

```python
print(generate_gpt2_words_cpu("Lyrics for a song of Taylor Swift starts with:\n\n"))
```

```python
print(generate_gpt2_words_cpu("Lyrics for a song of Eminem starts with:\n\n"))
```

```python
print(generate_gpt2_words_cpu("Lyrics for a song of Ariana Grande starts with:\n\n"))
```
