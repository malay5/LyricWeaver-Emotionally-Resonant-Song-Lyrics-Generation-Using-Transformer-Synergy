```python
import re
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
from collections import deque
from time import sleep

# Starting artist(s)
artists = ["edsheeran"]

# Max depth for BFS
max_depth = 200

# Base URL template for an artist's songs page
base_url_template = "https://www.song-list.net/{}/songs"

# To store songs per artist:
# { artist: { song_slug: { "url": full_url, "text": song_text } } }
songs_dict = {}

# BFS queue: each element is a tuple (artist, depth)
queue = deque([(artist, 0) for artist in artists])

# Set of visited artist names
visited = set(artists)

# Regex pattern to match an artist page URL (full or relative)
artist_page_pattern = re.compile(r"(?:https?://www\.song-list\.net|)(/([^/]+)/songs)")

seen={}
# Process the queue
while queue:
    current_artist, depth = queue.popleft()
    song_data = {}

    if current_artist not in seen:
        seen[current_artist] = 0
    else:
        seen[current_artist] += 1
        continue
    artist_url = base_url_template.format(current_artist)
    print(f"\nGoing to URL (Depth {depth}): {artist_url}")

    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
    req = Request(artist_url, headers=headers)

    try:
        html_page = urlopen(req)
        soup = BeautifulSoup(html_page, "html.parser")
    except Exception as e:
        print(f"Error fetching {artist_url}: {e}")
        sleep(5)
        continue

    # Initialize dictionary for current artist's songs


    # Look through all <a> tags on the page
    for a_tag in soup.find_all("a", href=True):
        href = a_tag["href"]
        # print("HREF:",href)
        text = a_tag.get_text(strip=True)


        # Look for song links specific to the current artist
        # They start with "/{current_artist}/" and contain "/tracks"
        if href.startswith(f"https://www.song-list.net/{current_artist}/") and "/tracks" in href:
            # print("FOUND FIRST")
            # Extract song slug from URL. Example: "/edsheeran/afirelove/tracks" -> "afirelove"
            parts = href.split("/")
            print("PARTS:",(parts),"Text",text)
            if len(parts) > 2:
                song_slug = parts[4]
                full_url = f"{href}"
                if len(text.strip())==0:
                  continue
                song_data[song_slug] = {"url": full_url, "text": text}

        # Look for links to other artist pages
        # This will catch both full and relative URLs that end with "/songs"
        match = artist_page_pattern.match(href)
        if match:
            # Extract the artist slug from the URL
            new_artist = match.group(2).lower()  # normalize to lower case
            # Skip if it's the current artist or already visited
            if new_artist != current_artist and new_artist not in visited:
                # print(f"  Found new artist: {new_artist} at URL: {href}")
                visited.add(new_artist)
                # Only add if we haven't reached max depth
                if depth + 1 <= max_depth:
                    queue.append((new_artist, depth + 1))

    songs_dict[current_artist] = song_data

    # Print some of the songs scraped for the current artist
    # print(f"Artist: {current_artist} - {len(song_data)} song(s) found.")
    # for song, data in list(song_data.items())[:5]:
    #     print(f"  Song Slug: {song}\n    Text: {data['text']}\n    URL: {data['url']}")

    # Sleep a bit to be polite to the server
    sleep(3)

print("\nBFS Crawl complete.")
print("Visited artists:")
print(visited)

# Optionally, print the full songs_dict or save it to a file
# For example:
import json
with open("songs_data.json", "w") as f:
    json.dump(songs_dict, f, indent=2)
```

```python
import re
import json
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
from collections import deque
from time import sleep
from tqdm import tqdm

# Starting artist(s)
# artists = ["edsheeran"]

# Max depth for BFS
max_depth = 200

# Base URL template for an artist's songs page
base_url_template = "https://www.song-list.net/{}/songs"

# To store songs per artist:
# { artist: { song_slug: { "url": full_url, "text": song_text } } }
songs_dict = {}

# BFS queue: each element is a tuple (artist, depth)
queue = deque([(artist, 0) for artist in artists])

# Set of visited artist names
visited = set(artists)

# Regex pattern to match an artist page URL (full or relative)
artist_page_pattern = re.compile(r"(?:https?://www\.song-list\.net|)(/([^/]+)/songs)")

seen = {}

# Wrap tqdm around queue for progress tracking
with tqdm(total=max_depth, desc="Processing Artists") as pbar:
    while queue:
        current_artist, depth = queue.popleft()
        song_data = {}

        if current_artist not in seen:
            seen[current_artist] = 0
        else:
            seen[current_artist] += 1
            continue

        artist_url = base_url_template.format(current_artist)
        print(f"\nGoing to URL (Depth {depth}): {artist_url}")

        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
        req = Request(artist_url, headers=headers)

        try:
            html_page = urlopen(req)
            soup = BeautifulSoup(html_page, "html.parser")
        except Exception as e:
            print(f"Error fetching {artist_url}: {e}")
            sleep(5)
            continue

        # Process all <a> tags on the page
        for a_tag in soup.find_all("a", href=True):
            href = a_tag["href"]
            text = a_tag.get_text(strip=True)

            # Look for song links specific to the current artist
            if href.startswith(f"https://www.song-list.net/{current_artist}/") and "/tracks" in href:
                parts = href.split("/")
                if len(parts) > 2:
                    song_slug = parts[4]
                    full_url = f"{href}"
                    if len(text.strip()) == 0:
                        continue
                    song_data[song_slug] = {"url": full_url, "text": text}

            # Look for links to other artist pages
            match = artist_page_pattern.match(href)
            if match:
                new_artist = match.group(2).lower()  # normalize to lower case
                if new_artist != current_artist and new_artist not in visited:
                    visited.add(new_artist)
                    if depth + 1 <= max_depth:
                        queue.append((new_artist, depth + 1))

        songs_dict[current_artist] = song_data

        # Save progress after each artist is processed
        with open("songs_data.json", "w") as f:
            json.dump(songs_dict, f, indent=2)

        # Update progress bar
        pbar.update(1)

        # Sleep to be polite to the server
        sleep(3)

print("\nBFS Crawl complete.")
print("Visited artists:", visited)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from time import sleep
import csv
import json
import re
```

```python
# file containing artists - songs mapping
songs_json = "songs_data.json"
songs_dict = {}

with open(songs_json) as file:
    songs_dict = json.load(file)
```

```python
# artist for which the lyrics need to be written
artist = "eminem"
songs = songs_dict[artist]
processed_songs = []

# preprocessing the songs name for scraping
for song in songs:
    numbers_in_brackets_removed = re.sub(r'\(.*\)',"",song)
    processed_song = re.sub(r'\W+', '', numbers_in_brackets_removed).lower()
    processed_songs.append(processed_song)

print(len(processed_songs))

# Removing duplicate songs
processed_songs = list(set(processed_songs))
print(len(processed_songs))
print(processed_songs[:20])
```
